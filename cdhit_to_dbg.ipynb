```python
from Bio import SeqIO
from itertools import groupby
from modified_dbg import assembly, testing_kmer_size
'''This code:
    1. read and extract the id from each cluster
    2. each cluster do DBG,
    3. output fasta for all contigs'''

#? read the CD-hit output file
#trim_result.clstr
with open("T_c_100k_filter_CDH.clstr", "r") as fcd:
    lines = fcd.readlines()
# print(lines)
#* each line in a list. <clstr>, <others...>, ...]


#? extract the read IDs and sequences from the lines
# read_dict = {}
all_dict = {}
read_list = []
clstr_id = None

for line in lines:
    # print(line)
    #if header -> obtain all the clstrs header
    if line[0] == ">": 
        #add current read_list into current cluster id
        if len(read_list) > 0:
            # print(read_list) #!Y
            all_dict[clstr_id] = read_list
            # print(all_dict)
        #initiate new read list
        read_list = []

        #new cluster id
        clstr_id = line.strip()[1:] #line.strip(), used to remove leading and trailing whitespaces from a string, [1:] means remove ">"

    else: #if not leader -> obtain all ids
        read_id = line.strip().split(",")[1]#split this line by "\t", take the 2nd element of line -> the IDs of the read, put in a dictionary
        read_id = read_id.split(" ")[1].replace(">", "").replace("...", "") #* remove those symbol, extract the id only
        # print(read_id)
        read_list.append(read_id)

        # print(read_id)
        # read_dict[read_id] = (clstr_id, line.strip().split("\t")[-1])
        #* each read is assigned to the latest cluster id
# print(all_dict)#! Y


#? from the dictionary, each key:value pair = each cluster id with reads
threshold = 0
for key in list(all_dict.keys()): #*the error, use .keys() instead of using all_dict only
    #filter cluster size smaller than a threshold value
    value = all_dict[key]

    #filter short cluster
    if len(value) < threshold:
        del all_dict[key]   
    
#return result dict after filtering
# for i in all_dict:
#     print(i, all_dict[i])

#? For each value = list of all id in dictionary, change it to the reads

#read in raw trimmed fasta
# FASTA file
fasta_file = "T_c_100k.fasta"

#obtain index from file
index = SeqIO.index(fasta_file, "fasta")
# for i in index:
#     print(i)

#(all clusters)obtain id list of each cluster
clstr_dict = {}
for i in all_dict:
    key = i #key

    #get the list of id
    id_list = all_dict[i] #value

    #(all id in each cluster)for each id in id list, obtain seq from file:
    seq_list = []
    for id in id_list:
        # print(id)
        if id in index:
            sequences = index[id].seq
            # print(sequences) #!Y can obtain each read
            # print("\n")
            # sequences = [index[id].seq for id in id_list if id in index] #if id in list in the file_index, obtain the .seq
            seq_list.append(str(sequences))            

    # print(seq_list) #all the reads from one cluster
    #each clusters sequences add to all clusters list
    clstr_dict[key] = seq_list

# print(len(clstr_dict))
# for i in clstr_dict:
    # print(len(clstr_dict[i]))
# print(len(clstr_dict))
# print(clstr_dict) #! Y: a dict of cluster id : list of all reads in each cluster


#?run dbg of each cluster, new dict -> cluster: reads list
#each cluster(list of reads), input to the dbg assembler
contigs_list = [] #*All final contigs from all cluster

for c in clstr_dict:
    # print(c)
    reads = clstr_dict[c]
    # print(reads)
    # #run dbg with fixed k-mer
    # i = 31
    # contigs, vertices, edges = assembly(reads, k=i)
    # print(contigs)
    
    #? cluster only has one read,skip the dbg
    if len(reads) == 1:
        contigs = reads[0]
        contigs_list.append(contigs)
        continue

    #?if not single reads, do dbg
    else:
        #run k-mer optimiser, return the best k-mer and its contigs
        max = 150  
        min = 10
        contigs, v, e = testing_kmer_size(min, max, reads)
        # print(v)
        # print(e)
        contigs_list.append(contigs)

    #? meanwhile, put the contigs to each a file with original reads and do comparison, for checking
    # sequences = clstr_dict[c]
    # with open(f"sequences_{c}.fasta", "w") as file: 
    #     #add the contigs of each cluster
    #     file.write(f">Final_contigs_{c}\n")
    #     file.write(contigs + "\n")

    #     #loop though each read
    #     for x in range(len(sequences)):
    #         # print((f">Cluster_{i}_Seqence{x}"))
    #         # print(sequences[x])
    #         file.write(f">Sequence_{x}_{c}\n") #print the header
    #         file.write(sequences[x] + "\n") #print the sequences

#?output a file for QC, of all final contigs
with open(f"sequences_all_contigs_.fasta", "w") as file: 
    i = 0
    for x in clstr_dict:
        #obtain the final contigs
        contigs = contigs_list[i]
        #cluster_ID
        x = x.replace(" ", "")
        #add the name of contigs as the header
        file.write(f">Final_contigs_{x}\n") 
        file.write(contigs + "\n")
        i += 1
```
